```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Distribution Graphs
#----->1.Distplot
#----->2.Jointplot
#----->3.Pairplot
#----->4.Rugplot

#Categorical Graphs
#----->1.Barplot
#----->2.Countplot
#----->3.Boxplot
#----->4.Violinplot

#Matrix Plot
#----->1.Heatmap
```

```python
#distplot
sns.distplot(tips['total_bill'])
```

```python
# jointplot(x->x-axis,y->y-axis and data->dataset_name,kind->graph)
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#pairplot->it only find the numerical columns in the database and plot the graph according to those columns
sns.pairplot(tips,hue='sex')
```

```python
#rugplot
sns.rugplot(tips['total_bill'])
```
