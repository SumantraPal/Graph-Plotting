```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
match=pd.read_csv('pandas/matches.csv')
delivery=pd.read_csv('pandas/deliveries.csv')
```

# 1.Pie Charts

```python
x=delivery[delivery['batsman']=='V Kohli']
y=x['batsman_runs'].value_counts()
plt.pie(y,labels=y.index,autopct="%0.2f%%")
plt.show
```

```python
def runPie(batsman_name):
    x=delivery[delivery['batsman']==batsman_name]
    y=x['batsman_runs'].value_counts()
    plt.pie(y,labels=y.index,autopct="%0.2f%%")
    plt.show()
    
```

```python
runPie('V Kohli') 
```

# 2.Line Chart

```python
#joining of two dataframes
mega=match.merge(delivery,left_on='id',right_on='match_id')
batsman_data1=mega[mega['batsman']=='V Kohli']
batsman_data2=mega[mega['batsman']=='RG Sharma']
performance1=batsman_data1.groupby('season').sum()['batsman_runs']
performance2=batsman_data2.groupby('season').sum()['batsman_runs']
performance1
performance2
plt.plot(performance1.index,performance1.values,color="green",marker='D',label="V Kohli")
plt.plot(performance2.index,performance2.values,color="red",marker='+',label="RG Sharma")
plt.xlabel('season')
plt.xlabel('Runs Scored')
plt.title('Batsman season by Season Performance')
plt.legend()
```

```python
def Comparator(batsman1,batsman2):
    mega=match.merge(delivery,left_on='id',right_on='match_id')
    batsman_data1=mega[mega['batsman']==batsman1]
    batsman_data2=mega[mega['batsman']==batsman2]
    performance1=batsman_data1.groupby('season').sum()['batsman_runs']
    performance2=batsman_data2.groupby('season').sum()['batsman_runs']
    
    plt.plot(performance1.index,performance1.values,color="green",marker='D',label=batsman1)
    plt.plot(performance2.index,performance2.values,color="red",marker='+',label=batsman2)
    plt.xlabel('season')
    plt.ylabel('Runs Scored')
    plt.title('Batsman season by Season Performance')
    plt.legend()
    plt.show()
```
